#**Vectorisation de chaque mot**

```python
# Tokenisation du texte
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
tokens
```

```python
# Entraîner le modèle word2vec
model = Word2Vec([tokens], min_count=1)
model
```

```python
# Vectorisation des mots et afficher ces mots et leurs vecteurs
for word in model.wv.index_to_key:
    vector = model.wv[word]
    print(f"{word}: {vector}")
```

**Calculer la similarite entre 2 mots**

```python
def similarity_calcul(word1, word2):
    if word1 in model.wv and word2 in model.wv:
        similarity = model.wv.similarity(word1, word2)
        return similarity
    else:
        return "ces mots ne sont pas dans le vocabulaire."
#tester la similarite avec "Marrakesh" et "cities"
score_similarity=similarity_calcul("Marrakesh","cities")
score_similarity
```

```python
#un autre test "Marrakesh" et "Morocco"
score_similarity=similarity_calcul('Morocco', 'Marrakech')
score_similarity
```

**Mots Contextuels:(Plus Similaires)**

```python
def Context_word(word, topn=5):
    if word in model.wv:
        context_words = model.wv.most_similar(word)
        return context_words
    else:
        return f"'{word}' n'est pas dans le vocabulaire."
#tester Pour "Marrakesh"
context_words = Context_word('Marrakesh')
context_words
```

```python
context_words = Context_word('Marrakech')
context_words
```
